# 특정 범위의 픽셀 변경하기

```python
import cv2
import matplotlib.pyplot as plt
import time
```

```python
image=cv2.imread('normal1.jpg')
```

```python
start_time=time.time()

#for문으로 개별적인 픽셀에 일일이 접근한다.
for i in range(0,100):
    for j in range(0, 100):
        image[i, j]=[255, 255, 255]    #선택된 픽셀에 지정된 값을 바꾼다
print("---%s seconds---"%(time.time() -start_time))
```

```python
start_time=time.time()
image[0:100, 0:100]=[0, 0, 0]   #범위를 지정한 다음에 해당 값을 동일한 값으로 바꾸는 방법
print("---%s seconds ---1" %(time.time() -start_time))
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

# ROI (Region of Interest)

```python
#Numpy Slicing: ROI 처리 가능
cat=cv2.imread('cat.jpg')
roi=cat[500: 800, 300: 600]    #복사하고자 하는 픽셀의 부분

#ROI 단위로 이미지 복사하기
cat[0:300, 0:300]=roi         #붙여 넣고자 하는 픽셀의 부분
plt.imshow(cv2.cvtColor(cat, cv2.COLOR_BGR2RGB))
plt.show()
```

# 픽셀별로 색상 다루기

```python
cat[:,:,2]=0   #(모든 픽셀에 대해서 index 2에 해당하는)빨간에 해당하는 칙셀 값들을 전부 0으로 만들기

plt.imshow(cv2.cvtColor(cat, cv2.COLOR_BGR2RGB))
plt.show()
```
