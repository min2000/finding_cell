이미지를 읽어서 numpy 객체로 만들기

**cv2.imread(file_name, flag): 이미지를 읽어 Numpy 객체로 만드는 함수**
~~~
-file_name: 읽고자 하는 이미지 파일
-flag: 이미지를 읽는 방법 설정
IMREAD_COLOR: 이미지를  color로 읽고 투명한 부분은 무시
IMREAD_GRAYSCALE: 이미지를 gratscale로 읽기
IMREAD_UNCHANGED: 이미지를 color로 읽고 투명한 부분도 읽기(Alpha)
~~~
**cv2.imshow(title, image): 특정한 이미지를 화면에 출력**
~~~
title: 윈도우 창의 제목
image: 출력할 이미지 객체
~~~
**cv2.imwrite(file_name, image): 특정한 이미지를 파일로 저장하는 함수**
~~~
-file_name: 저장할 이미지 파일 이름
-image: 저장할 이미지 객체
~~~
**cv2.waitKey(time): 키보드 입력을 처리하는 함수**
~~~
time: 입력 대기 시간 (무한 대기: 0)
변환 값: 사용자가 입력한 Ascii Code (ESC:27)
~~~

~~~
color_imag=cv2.imread('normal1.jpg', cv2.IMREAD_COLOR)    #색이 포함된 이미지 읽기
cv2.imshow('colorImage', color_imag)                      #이미지 보여주기
cv2.waitKey(0)                                            #이미지를 보여준 후에 꺼지지 않게 하기
cv2.imwrite('result1.png', color_imag)                    #이미지를 다른 형태로 저장
~~~

~~~
gray_imag=cv2.cvtColor(img_basic, cv2.COLOR_BGR2GRAY)  
cv2.imshow('grayImage', gray_imag)          #흑백 이미지 출력
cv2.waitKey(0)
cv2.imwrite('result2.png', gray_imag)
~~~

```python
import cv2
import matplotlib.pyplot as plt
```

```python
color_imag=cv2.imread('normal1.jpg', cv2.IMREAD_COLOR) 
plt.imshow(cv2.cvtColor(color_imag, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
color_imag=cv2.cvtColor(color_imag, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(color_imag, cv2.COLOR_GRAY2RGB))
plt.show()
```
